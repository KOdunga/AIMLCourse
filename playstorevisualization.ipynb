```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option('display.float_format', lambda x:'%.2f' %x) # only show 3 decimal places for floats
```

```python
data = pd.read_csv('/content/Play_Store_2.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe(include = 'all').T
```

```python
data.isnull().sum()
```

```python
data['Rating'] = data['Rating'].fillna(value = data['Rating'].mean())
```

```python
data['Reviews'] = data['Reviews'].fillna(value = data['Reviews'].mean())
```

```python
data.isnull().sum()
```

```python
sns.histplot(data = data, x ='Rating');
```

```python
sns.boxplot(data = data, x = 'Rating');
```

```python
sns.histplot(data = data, x ='Reviews')
plt.show()
sns.boxplot(data = data, x = 'Reviews')
plt.show()
```

```python
plt.figure(figsize = (15,7))
sns.countplot(data = data, x = 'Category')
plt.xticks(rotation = 90);
```

```python
sns.countplot(data = data, x = 'Paid/Free');
```

```python
sns.countplot(data = data, x = 'Content Rating');
```

```python
sns.countplot(data = data[data['Paid/Free']=='Free'], x = 'Ad Supported');
```

```python
sns.displot(data= data, x = 'Price', kind = 'kde');
```

```python
sns.histplot(data = data, x = 'Price');
```

```python
plt.figure(figsize = (15,7))
sns.boxplot(data=data, x = 'Category', y = 'Rating')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (15,7))
sns.lineplot(data = data, x = 'Category', y = 'Reviews',ci=False,estimator = 'sum')
plt.xticks(rotation = 90);
```

```python
plt.figure(figsize = (15,7))
sns.countplot(data = data[data['Paid/Free'] =='Free'], x = 'Category', hue = 'Ad Supported')
plt.xticks(rotation = 90);
```

```python
plt.figure(figsize = (10,5))
sns.heatmap(data.corr(), annot= True, cmap = 'Spectral', vmin = -1, vmax=1);
```

```python
sns.lmplot(data = data, x = 'Installs', y = 'Reviews', ci = None);
```

```python
sns.pairplot(data = data, diag_kind = 'kde');
```

```python
numeric_columns = ['Rating','Reviews','Size','Installs','Price']
plt.figure(figsize = (15,12))

for i, variable in enumerate(numeric_columns):
  plt.subplot(4,4, i+1)
  plt.boxplot(data[variable], whis = 1.5)
  plt.tight_layout()
  plt.title(variable)
plt.show()
```
