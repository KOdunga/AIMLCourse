## **Analysis of Tips in a Hotel**

Import packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option('display.float_format', lambda x:'%.2f' %x) # only show 3 decimal places for floats
```

Import data

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

Understand Data

```python
tips.shape
```

- There are 244 entries in the table, with 8 column descriptions

```python
tips.info()
```

```python
tips.describe(include = 'all').T
```

- We have 244 non null entries
- Average bill is around 20 dollars
- Max bill paid is 50 Dollars
- Least bill paid is 3 dollars
- Average tip amount is 3 dollars
- Max tip paid is 10 dollars, min is 1 dollar
- A table has on average 2 people, max being 6 and min is 1
- Most visited day is Sartuday
- Most visited time is Dinner
- Most guests are male

```python
tips.isnull().sum()
```

- There is no missing data

Univariate Analysis
- Use histplot and boxplot on numerical variables
- Cant work on categorical variables

```python
sns.histplot(data = tips, x = 'size')
plt.show()
sns.boxplot(data = tips, x = 'size')
plt.show()
```

- Size has outliers
- Mostly two people per table

```python
sns.histplot(data = tips, x = 'total_bill')
plt.show()
sns.boxplot(data = tips, x = 'total_bill')
plt.show()
```

- Bills are largely around 10 to 30 dollars
- Some people pay large amounts of upto 50 dollars

```python
sns.histplot(data = tips, x = 'tip')
plt.show()
sns.boxplot(data = tips, x = 'tip')
plt.show()
```

```python
sns.countplot(data = tips, x = 'time');
```

```python
sns.countplot(data = tips, x = 'day', hue = 'sex');
```

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill', jitter = True);
```

```python
sns.swarmplot(data = tips, x = 'day', y = 'total_bill');
```

- Male visitors dominate on most days except on Thursday
- Sartuday and Sunday have most visitors

Bivariate Analysis

```python
sns.lineplot(data = tips, x = 'day', y='tip', ci =False, estimator = 'sum');
```

```python
sns.lineplot(data = tips, x = 'day', y='total_bill', ci =False, estimator = 'sum');
```

- Sartuday and Sunday, having most number of visitors, also has the highest tipping and total bills paid

```python
sns.scatterplot(data = tips, x = 'tip', y = 'total_bill');
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hex');
```

```python
sns.lmplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', ci =False);
```

```python
sns.lmplot(data = tips, x = 'total_bill', y = 'tip', col = 'smoker', ci =True);
```

```python
sns.lmplot(data = tips, x = 'total_bill', y = 'tip', col = 'day', hue ='sex');
```

Multivariate Analysis

```python
num_cols = ['tip', 'total_bill','size']
sns.pairplot(data = tips[num_cols], kind = 'kde')
plt.show()
```

```python
plt.figure(figsize = (15,7))
sns.heatmap(tips[num_cols].corr(), annot= True, cmap = 'Spectral', fmt = '.2f');
```

- We see a high correlation across the variables of tip, total bill and number of people per table

```python
plt.figure(figsize =(15,7))
sns.boxplot(x = tips['day'], y = tips['total_bill'])
sns.set_style("darkgrid")
plt.ylabel('Total Bill')
plt.xlabel('Day')
plt.show()
```

- In all days, there is payment of high total bills by customers. 
- The average bill per day is almost the same across all days.
