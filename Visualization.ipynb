```python
# import the important packages
import pandas as pd  # library used for data manipulation and analysis
import numpy as np  # library used for working with arrays
import matplotlib.pyplot as plt  # library for plots and visualizations
import seaborn as sns  # library for visualizations

%matplotlib inline

import scipy.stats as stats  # this library contains a large number of probability distributions as well as a growing library of statistical functions
```

```python
automobile = pd.read_csv('/content/Automobile.csv')
automobile.info()
automobile.head()
```

```python
automobile.shape
```

```python
automobile.isna().sum()
```

```python
automobile.describe(include= 'all').T
```

**Histogram**

```python
sns.histplot(data = automobile, x = 'price'); # Add semi colon to remove the memory information
```

```python
plt.title('Price')
plt.xlim(3000,50000)
plt.ylim(0,70)
plt.xlabel('price of cars')
plt.ylabel('frequency')
sns.histplot(data = automobile, x = 'price', color = 'orange');
```

```python
sns.histplot(data = automobile, x = 'price', bins = 5);# specify number of bins
```

```python
sns.histplot(data = automobile, x = 'price', binwidth = 1000); #specify binwidth
```

**Finding optimal number of bins**
 - binwidth = (2*IQR)/cuberoot(n)
 - bins = range/binwidth


```python
#KDE for trending. Multiplies the frequency of the bin and its probability
sns.histplot(data = automobile, x = 'price', kde = True); #Kernel Density estimation
```

```python
sns.histplot(data = automobile, x = 'curb_weight', kde = True);
```

```python
sns.histplot(data = automobile, x = 'price', hue = 'body_style'); #Use hue to describe distribution within groups
```

```python
#Separating the various body type plots with different subplots using Facetgrid
g = sns.FacetGrid(automobile, col = 'body_style')
g.map(sns.histplot,'price');
```

**BoxPlots**

```python
sns.boxplot(automobile, x ='curb_weight');
```

```python
#drawing multiple box plots on the same graph

plt.figure(figsize=(10,7))
sns.boxplot(data= automobile, x = 'body_style', y ='price');
```

## **Plotly**
- Open source Python graphing library for building beautiful and interactive visualizations

```python
#Install Plotly. Ccmes installed in google colab
!pip install plotly
```

```python
import plotly.express as px
```

```python
his = px.histogram(automobile, x = 'price')
his.show()
#Zoom, hover, move graph using mouse
#Double click on graph to restore normal
```

```python
#Bar Plot
bar = px.bar(automobile, x = 'peak_rpm',y = 'horsepower')
bar.show()
```

```python
#Scatter Plot
scat = px.scatter(automobile, x = 'price', y = 'engine_size')
scat.show()
```

```python
#Box plot with underlying data
fig = px.box(automobile, x = 'fuel_type',y = 'horsepower', points = 'all')
fig.show()
```

```python
#3D Scatter Plot
fig_3d = px.scatter_3d(automobile, x = 'fuel_type',y = 'horsepower',z = 'price',color = 'horsepower')
fig_3d.show()
```

```python
#Saving the file
fig_3d.write_html('scatter_3d.html')
```
