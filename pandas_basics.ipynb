To drop a column

```python
store_data.drop('CustomerID', axis = 1) #returns a new data frame that has all data except the dropped column. axis = 1 indicates we are checking in the columns
```

Combining Data Frames
- Concatenation
- Join - combine using the index to identify common entries
- Merging - combine using a columns values to identify common entries

```python
calorie_data
```

```python
calorie_data_new = pd.DataFrame({'day':['day1','day2','day4'],
                             'kms':[30,5,4]
                             })
calorie_data_new
```

```python
pd.concat([calorie_data,calorie_data_new], axis = 0) #concatenate by row
```

```python
pd.concat([calorie_data,calorie_data_new], axis = 1) #concatenate by column
```

Merge
- How = Outer Means union join
- How = Inner Means and intersection join
- How = Left Means use the first table entries compare with the second
- How = Right Means use the entries of the second table as key
- On = 'CustomerId' means use customer ID as the primary key for merging

```python
pd.merge(calorie_data,calorie_data_new, how = 'outer', on = 'day')
```

```python
pd.merge(calorie_data,calorie_data_new, how = 'inner', on = 'day')
```

```python
pd.merge(calorie_data,calorie_data_new, how = 'left', on = 'day')
```

```python
pd.merge(calorie_data,calorie_data_new, how = 'right', on = 'day')
```

Join
- Instead of 'on =customer id ' use the index
- Syntax is also different from merge
- firstdataframe.join(secondframe)

```python
data_quarters = pd.DataFrame({'Q1':[101,102,103],
                              'Q2':[201,202,203]},
                             index = ['I0','I1','I2'])

data_quarters_new = pd.DataFrame({'Q3':[301,302,303],
                                  'Q4':[401,402,403]},
                             index = ['I0','I1','I3'])
```

```python
data_quarters
```

```python
data_quarters_new
```

```python
data_quarters.join(data_quarters_new, how = 'right') #outer, inner, right and left work the same as merge
```

```python
data_quarters.join(data_quarters_new, how = 'outer')
```

```python
data_quarters.join(data_quarters_new, how = 'inner')
```

```python
data_quarters.join(data_quarters_new, how = 'left')
```
